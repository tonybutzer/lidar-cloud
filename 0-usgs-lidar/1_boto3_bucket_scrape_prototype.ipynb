```python
import boto3
```

```python
def _return_prefixes(bucket):
    PREFIXES=[]
    s3_client = boto3.client('s3')

    response = s3_client.list_objects_v2(Bucket=bucket, Delimiter = '/')

    for prefix in response['CommonPrefixes']:
         #print(prefix['Prefix'][:-1])
        PREFIXES.append(prefix['Prefix'][:-1])
    return PREFIXES
```

```python
bucket='pangeo-lidar-demo'
```

```python
prefixes = _return_prefixes(bucket)
```

```python
def _return_s3_object_list(bucket):
    image_list=[]
    s3 = boto3.resource('s3')
    some_path = bucket
    my_bucket = s3.Bucket(some_path)
    for my_bucket_object in my_bucket.objects.all():
        #print(my_bucket_object)
        #print(my_bucket_object.key)
        image_list.append(my_bucket_object.key)
    return image_list
```

```python
objects = _return_s3_object_list(bucket)
```
